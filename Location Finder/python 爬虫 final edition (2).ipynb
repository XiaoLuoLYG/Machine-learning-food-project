```python
import requests
import re
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
from time import sleep
import random
import json
import io
import os

path = 'd:/我的文档/Food Project/FoodProject/dianpingData/data/'
fnames = os.listdir(path)

fpaths = []
for name in fnames:
    fpath = path + name
    fpaths.append(fpath)
print(fpaths)
```

```python
def getFile(x):
    filepath = fpaths[x]
    return filepath


def fileread(x):
    filepath = getFile(x)

    f = io.open(filepath, encoding = 'utf-8')
    f1 = ""
    for line in f:
        f1 = f1 + line
    return f1
```

```python
rating = []
name = []
shopid = []

for a in range(len(fpaths)):
    f1 = fileread(a)
    soup = BeautifulSoup(f1, 'html')

    #find rating by looking for star score in the source code
    for i in range(30, 55, 5):
        character = "star_score score_{} star_score_sml".format(i)
        character1= '[<div class="star_score score_{} star_score_sml"'.format(i)
        r = soup.find_all('div', class_=character)
        string = str(r)
        t = string.split(",")
        for j in range(len(t)):
            t1 = t[j].strip(character1)
            t2 = t1.strip('[</div>]')
            rating.append(t2)

    #find shop name by looking for title
    n = soup.find_all('img')

    NameString = str(n)
    title = re.findall(r'title\=\".*?\"',NameString)

    for i in range(len(title)):
        t = eval(title[i].split('=')[1])
        name.append(t)

    #find shop id by looking for id data
    shoplink = soup.find_all('a')
    shoplink = str(shoplink)
    shopid1 = re.findall(r'data-shopid\=\".*?"', shoplink)
    shopid2 = []
    for item in shopid1:
        if not item in shopid2:
            shopid2.append(item)

    for i in range(len(shopid2)):
        s = eval(shopid2[i].split('=')[1])
        shopid.append(s)

print(rating)
print(name)
print(shopid)
print(len(rating))
print(len(name))
print(len(shopid))
```
